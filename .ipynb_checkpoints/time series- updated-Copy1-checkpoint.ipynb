# Grid Search for Arima

```python
def arima_grid_search(train_data, test_data):
    # Define the p, d, q ranges
    p = range(0, 5)
    d = range(0, 2)
    q = range(0, 5)
    
    # Initialize variables to track the best model
    best_score, best_cfg, best_predictions = float("inf"), None, None
    
    # Extract the temperature series for training and testing
    train_values = train_data['celsius_temperature']
    test_values = test_data['celsius_temperature']
    
    # Grid search over p, d, q
    for param_p in p:
        for param_d in d:
            for param_q in q:
                try:
                    # Fit ARIMA model
                    model = ARIMA(train_values, order=(param_p, param_d, param_q))
                    model_fit = model.fit()
                    
                    # Forecast
                    predictions = model_fit.forecast(steps=len(test_values))
                    
                    # Calculate the error
                    error = mean_squared_error(test_values, predictions)
                    
                    # Update the best model if error improves
                    if error < best_score:
                        best_score, best_cfg = error, (param_p, param_d, param_q)
                        best_predictions = predictions
                    
                    print(f"ARIMA({param_p},{param_d},{param_q}) MSE={np.sqrt(error)}")
                
                except Exception as e:
                    print(f"Error with ARIMA({param_p},{param_d},{param_q}): {e}")
                    continue
    
    print(f"Best ARIMA({best_cfg}) MSE={np.sqrt(best_score)}")
    
    return best_predictions  # Return the best predictions

# Call the function and get the best predictions
best_predictions = arima_grid_search(train_data, test_data)
```

# Bayesian Optimization for ARIMA

```python
from bayes_opt import BayesianOptimization
```

```python
def arima_bayesian_optimization(train_data, test_data):
    # Define the objective function for RMSE
    def arima_rmse(p, d, q):
        # Train the ARIMA model with the current p, d, q values
        model = ARIMA(train_data['celsius_temperature'], order=(int(p), int(d), int(q)))
        model_fit = model.fit()
        
        # Generate predictions for the test data
        predictions = model_fit.forecast(steps=len(test_data))
        
        # Calculate the RMSE
        rmse = np.sqrt(mean_squared_error(test_data['celsius_temperature'], predictions))
        
        # Return the negative RMSE for minimization
        return -rmse

    # Perform Bayesian optimization
    optimizer = BayesianOptimization(
        f=arima_rmse, 
        pbounds={'p': (0, 4), 'd': (0, 1), 'q': (0, 4)},  # Search space for ARIMA parameters
        random_state=42
    )
    
    # Maximize the objective function (minimizing RMSE by returning negative values)
    optimizer.maximize(init_points=10, n_iter=30)
    
    # Retrieve the best parameters
    best_params = optimizer.max['params']
    print(f"Optimal ARIMA parameters: p={int(best_params['p'])}, d={int(best_params['d'])}, q={int(best_params['q'])}")

# Call the Bayesian optimization function with your train and test data
arima_bayesian_optimization(train_data, test_data)
```

# SARIMAX with Grid Search

```python
def sarimax_grid_search(train_data, test_data):
    p = range(0, 5)
    d = range(0, 2)
    q = range(0, 5)
    P = range(0, 5)
    D = range(0, 2)
    Q = range(0, 5)
    s = 8  # Adjusted seasonality for daily cycle in 3-hour intervals (you can change this based on your needs)

    best_score, best_cfg, best_predictions = float("inf"), None, None
    
    # Extract the temperature series for training and testing
    train_values = train_data['celsius_temperature']
    test_values = test_data['celsius_temperature']
    
    # Perform grid search over p, d, q, P, D, Q
    for param_p in p:
        for param_d in d:
            for param_q in q:
                for param_P in P:
                    for param_D in D:
                        for param_Q in Q:
                            try:
                                # Fit the SARIMAX model
                                model = SARIMAX(train_values, 
                                                order=(param_p, param_d, param_q), 
                                                seasonal_order=(param_P, param_D, param_Q, s))
                                
                                model_fit = model.fit(disp=False)
                                
                                # Forecast for the test data period
                                predictions = model_fit.forecast(steps=len(test_values))
                                
                                # Calculate the RMSE
                                error = mean_squared_error(test_values, predictions)
                                
                                # Update best model if we find a lower error
                                if error < best_score:
                                    best_score = error
                                    best_cfg = (param_p, param_d, param_q, param_P, param_D, param_Q)
                                    best_predictions = predictions
                                
                                print(f"SARIMAX({param_p},{param_d},{param_q})x({param_P},{param_D},{param_Q},{s}) MSE={np.sqrt(error)}")
                            
                            except Exception as e:
                                print(f"Error with SARIMAX({param_p},{param_d},{param_q})x({param_P},{param_D},{param_Q},{s}): {e}")
                                continue

    print(f"Best SARIMAX{best_cfg} MSE={np.sqrt(best_score)}")
    
    return best_predictions  # Return the best predictions

# Call the grid search function
best_sarimax_predictions = sarimax_grid_search(train_data, test_data)
```

### Bayesian Optimization for SARIMAX

```python
def sarimax_bayesian_optimization(train_data, test_data, m):
    # Define the objective function for RMSE
    def sarimax_rmse(p, d, q, P, D, Q):
        # Fit SARIMAX model with given (p, d, q) and (P, D, Q) parameters
        model = SARIMAX(train_data['celsius_temperature'], 
                        order=(int(p), int(d), int(q)), 
                        seasonal_order=(int(P), int(D), int(Q), m))
        
        model_fit = model.fit(disp=False)
        
        # Forecast for the test data period
        predictions = model_fit.forecast(steps=len(test_data))
        
        # Calculate the RMSE
        rmse = np.sqrt(mean_squared_error(test_data['celsius_temperature'], predictions))
        
        # Return the negative RMSE to minimize it
        return -rmse

    # Perform Bayesian optimization
    optimizer = BayesianOptimization(
        f=sarimax_rmse, 
        pbounds={'p': (0, 5), 'd': (0, 2), 'q': (0, 5), 'P': (0, 5), 'D': (0, 2), 'Q': (0, 5)}, 
        random_state=42
    )
    
    # Maximize the objective function (minimizing RMSE by returning negative values)
    optimizer.maximize(init_points=10, n_iter=30)
    
    # Retrieve the best parameters
    best_params = optimizer.max['params']
    print(f"Optimal SARIMAX parameters: p={int(best_params['p'])}, d={int(best_params['d'])}, q={int(best_params['q'])}, "
          f"P={int(best_params['P'])}, D={int(best_params['D'])}, Q={int(best_params['Q'])}")
    
    return best_params

# Call the Bayesian optimization function with appropriate seasonality (m=8 for daily cycle in 3-hour intervals)
best_params = sarimax_bayesian_optimization(train_data, test_data, m=8)
```

```python
from prophet import Prophet
```

```python
df = pd.read_csv('temp_series_data.csv')
```

```python
# Ensure 'timestamp' is in datetime format with the correct format
data['timestamp'] = pd.to_datetime(data['timestamp'], format='%d/%m/%Y %H:%M')
```

```python
# Rename relevant columns to match Prophet's expectations
df = df[['timestamp', 'celsius_temperature']]
df.columns = ['ds', 'y']  # Prophet expects 'ds' for date and 'y' for values
```

```python
# Convert 'ds' to datetime and ensure 3-hour frequency
df['ds'] = pd.to_datetime(df['ds'], dayfirst=True)
df = df.set_index('ds').resample('3H').mean().reset_index()
```

```python
# Plot the time series data
df.plot(x='ds', y='y', figsize=(12,5)).autoscale(axis='x', tight=True)
plt.show()
```

This graph shows the original time series data, which is the recorded temperature over time.

This is purely the observed data and helps you visually understand the behavior of the temperature over time before applying a forecasting model for understanding trends and seasonality.

```python
# Split the data into training and testing sets
train, test = df.iloc[:int(len(df)*0.8)], df.iloc[int(len(df)*0.8):]
```

```python
# Create and train the model
m = Prophet().fit(train)
```

```python
# Create a future dataframe (predicting 12 periods ahead, at 3-hour intervals)
future = m.make_future_dataframe(periods=12, freq='3H')
```

```python
# Generate forecasts
forecast = m.predict(future)
```

```python
# Plot the forecast
ax = forecast.plot(x='ds', y='yhat', label='Predictions', figsize=(12,5), color='red')
test.plot(x='ds', y='y', label='True test data',legend=True, ax=ax, color='black');
```

This graph shows the forecasted values (in red) versus the actual test data (in black).

This graph compares how well the model predicts future temperatures compared to actual observations.

If the red line closely follows the black dots, it means the model is doing a good job of forecasting future temperature.

Wide shaded areas indicate greater uncertainty in the model's prediction.

If the black dots fall outside the shaded area, it means the model is not capturing the actual temperatures well, which could indicate the need for model tuning.


```python
# Evaluation
from statsmodels.tools.eval_measures import rmse

# Get predictions for the last 12 periods (test set)
predictions = forecast.iloc[-12:]['yhat']

# Calculate RMSE
rmse_val = rmse(predictions, test['y'].head(12))
print(f'RMSE: {rmse_val}')
```

```python
test.mean()
```

```python
# when running cross validation metric, we need to define 3 things:
# 1. initial training period
# 2. the period length that we're going to perform cross validation for
# 3. the horizon of period for each fold

# Prophet built-in cross-validation diagnostics
from prophet.diagnostics import cross_validation, performance_metrics
from prophet.plot import plot_cross_validation_metric

# Initial, period, and horizon for cross-validation (adjusted for 3-hour data)
initial = '30 days'  # Smaller initial window to account for the dataset size
period = '15 days'    # Time between cross-validation cutoffs
horizon = '30 days'   # Forecasting 15 days into the future
```

```python
# Cross-validation
df_cv = cross_validation(m, initial=initial, period=period, horizon=horizon)
df_cv.head()
```

```python
performance_metrics(df_cv)
# this tells us for each horizon period, what the stats or evaluation metrics are.
# mape = mean absolute percentage error
```

```python
# Plot cross-validation metrics
plot_cross_validation_metric(df_cv, metric='rmse')
plt.show()
```

This graph shows the Root Mean Square Error (RMSE) over time, with blue and grey dots representing the accuracy of the model at different forecast horizons (future time points).

It shows how the model's error changes as it tries to predict further into the future. The further you go into the future, the harder it becomes for the model to maintain accuracy.

Lower RMSE values (blue line closer to the bottom) indicate better performance, meaning the model's predictions are closer to the actual values.

If the blue line increases over time, it means the forecast becomes less accurate as it predicts further into the future.

Grey dots represent individual forecast errors at each time point, providing insight into the variability of the model's performance.
